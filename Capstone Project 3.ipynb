# Data Analysis

Countplot Churn

```python
plt.figure(figsize = (8,4))
ax = sns.countplot(x = df["Churn"], palette = {"Yes": "orange", "No": "blue"})
ax.bar_label(ax.containers[0], color = "black")
plt.title("Distribution Churn")
plt.ylabel("Number of Customers")
plt.show()
```

Countplot based on Customer Demographics

```python
features = ["gender","Partner","Dependents"]
plt.figure(figsize = (18,6))
for i in range(0,len(features)):
    plt.subplot(1, len(features), i+1)
    ax = sns.countplot(x = df[features[i]], palette = {"Yes": "Orange", "No": "Blue", "Male": "Orange", "Female": "Blue"})
    ax.bar_label(ax.containers[0], color = "Black")
    plt.title("Distribution " + features[i])
    plt.ylabel("Number of Customers")
    plt.tight_layout()
```

Countplot Based on Service Type

```python
features = ["PhoneService","MultipleLines","InternetService","OnlineSecurity","OnlineBackup","DeviceProtection","TechSupport","StreamingTV","StreamingMovies"]
plt.figure(figsize = (12,16))
for i in range(0,len(features)):
    plt.subplot(4, len(features)//3, i+1)
    ax = sns.countplot(x = df[features[i]], palette = {"Yes": "Orange", "No": "Blue", "No phone service": "red", "DSL": "Red", "Fiber optic": "Orange", "No Internet": "Red"})
    ax.bar_label(ax.containers[0], color = "Black")
    plt.title("Distribution " + features[i])
    plt.ylabel("Number of Customers")
    plt.tight_layout()
    
```

Countplot based on Payment Type

```python
features = ["Contract","PaperlessBilling","PaymentMethod"]
plt.figure(figsize = (12,6))
for i in range(0, len(features)):
    plt.subplot(1, len(features), i+1)
    ax = sns.countplot(x = df[features[i]], palette = {"Month-to-month": "Blue", "One year": "Orange", "Two year": "Green", "Yes": "Blue", "No": "Orange", "Electronic check": "Blue", "Mailed check": "Orange", "Bank transfer (automatic)": "Green", "Credit card (automatic)": "Red"})
    ax.bar_label(ax.containers[0], color = "Black")
    plt.title("Distribution " + features[i])
    plt.ylabel("Number of Customers")
    plt.xticks(rotation = 25)
    plt.tight_layout()
```

Churn Barplot based on Customer Demographics

```python
features = ["gender","Partner","Dependents"]
plt.figure(figsize = (10,8))
for i in range(0,len(features)):
    Churn_Customer = df.groupby(features[i],as_index = False)["Churn"].value_counts()
    plt.subplot(3, len(features)//2, i+1)
    ax = sns.barplot(x = features[i], palette = {"Yes": "Orange", "No": "Blue", "Male": "Orange", "Female": "Blue"}, y = "count", hue="Churn", data = Churn_Customer)
    ax.bar_label(ax.containers[0], color = "Black")
    ax.bar_label(ax.containers[1], color = "Black")
    plt.title("Churn By " + features[i])
    plt.ylabel("Number of Customers")
    plt.tight_layout()
```

Churn Barplot based on Service Type

```python
features = ["PhoneService","MultipleLines","InternetService","OnlineSecurity","OnlineBackup","DeviceProtection","TechSupport","StreamingTV","StreamingMovies"]
plt.figure(figsize = (14,16))
for i in range(0,len(features)):
    Churn_Service = df.groupby(features[i],as_index = False)["Churn"].value_counts()
    plt.subplot(4, len(features)//3, i+1)
    ax = sns.barplot(x = features[i], palette = {"Yes": "Orange", "No": "Blue", "No phone service": "red", "DSL": "Red", "Fiber optic": "Orange", "No Internet": "Red"}, y = "count", hue="Churn", data = Churn_Service)
    ax.bar_label(ax.containers[0], color = "Black")
    ax.bar_label(ax.containers[1], color = "Black")
    plt.title("Churn By " + features[i])
    plt.ylabel("Number of Customers")
    plt.tight_layout()
```

Churn Barplot based on Payment Type

```python
features = ["Contract","PaperlessBilling","PaymentMethod"]
plt.figure(figsize = (12,6))
for i in range(0, len(features)):
    Churn_Payment = df.groupby(features[i],as_index = False)["Churn"].value_counts()
    plt.subplot(1, len(features), i+1)
    ax = sns.barplot(x = features[i], palette = {"Month-to-month": "Blue", "One year": "Orange", "Two year": "Green", "Yes": "Blue", "No": "Orange", "Electronic check": "Blue", "Mailed check": "Orange", "Bank transfer (automatic)": "Green", "Credit card (automatic)": "Red"}, y = "count", hue="Churn", data = Churn_Payment)
    ax.bar_label(ax.containers[0], color = "Black")
    ax.bar_label(ax.containers[1], color = "Black")
    plt.title("Chuen By " + features[i])
    plt.ylabel("Number of Customers")
    plt.xticks(rotation = 25)
    plt.tight_layout()
```
